```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, array_to_latex
```

```python
circuit = QuantumCircuit(1, 0)
circuit.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend = backend).result()
state_vector = result.get_statevector()
plot_bloch_multivector(state_vector)
```

```python
state_vector.draw('qsphere')
```

```python
state_vector = Statevector.from_int(0, 2**1)
state_vector = state_vector.evolve(circuit)
state_vector.draw('latex')
```

```python
array_to_latex(state_vector)
```
