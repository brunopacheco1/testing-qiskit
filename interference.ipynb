```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import numpy as np

simulator = Aer.get_backend('qasm_simulator')

def print_circuit(rotation: float):
    print("=============================")
    circuit = QuantumCircuit(2)
    circuit.h(0)
    circuit.crx(rotation * 2 * np.pi, 0, 1)
    state_vector = Statevector.from_instruction(circuit)
    circuit.measure_all()
    display(circuit.draw('mpl'))
    result = execute(circuit, backend=simulator, shots=10000).result()
    display(plot_bloch_multivector(state_vector))
    display(state_vector.draw('qsphere'))
    display(plot_histogram(result.get_counts()))
```

```python
splits = 4
for i in range(splits+1):
    print_circuit(i/(splits))

print_circuit(1.5)
```
