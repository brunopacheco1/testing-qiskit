```python
from qiskit import QuantumCircuit, Aer
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, array_to_latex
import numpy as np

%matplotlib inline

simulator = Aer.get_backend('qasm_simulator')

def print_circuit(circuit: QuantumCircuit):
    display(circuit.draw('mpl'))
    state_vector = Statevector.from_instruction(circuit)
    display(plot_bloch_multivector(state_vector))
    display(state_vector.draw('qsphere'))
    display(state_vector.draw('latex'))
    display(array_to_latex(state_vector))
```

```python
circuit = QuantumCircuit(1)
print_circuit(circuit)
```

```python
circuit = QuantumCircuit(1)
circuit.x(0)
print_circuit(circuit)
```

```python
circuit = QuantumCircuit(1)
circuit.y(0)
print_circuit(circuit)
```

```python
circuit = QuantumCircuit(1)
circuit.z(0)
print_circuit(circuit)
```

```python
circuit = QuantumCircuit(1)
circuit.h(0)
print_circuit(circuit)
```
