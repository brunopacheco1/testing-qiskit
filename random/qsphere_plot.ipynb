```python
from qiskit import QuantumCircuit
import numpy as np
from qiskit.quantum_info import Statevector
```

```python
number_of_bits = 2
circuit = QuantumCircuit(number_of_bits, number_of_bits)
circuit.h(0)
circuit.cx(0,1)
circuit.x(1)
state_vector = Statevector.from_instruction(circuit)
state_vector.draw('qsphere')
```

```python
number_of_bits = 6
circuit = QuantumCircuit(number_of_bits, number_of_bits)
for i in range(number_of_bits):
    circuit.h(i)
    circuit.rz(2 * np.pi / number_of_bits, i)
state_vector = Statevector.from_instruction(circuit)
state_vector.draw('qsphere')
```

```python
circuit.draw('mpl')
```

```python
number_of_bits = 4
circuit = QuantumCircuit(number_of_bits, number_of_bits)
circuit.h(0)
for i in range(number_of_bits - 1):
    circuit.cx(0, i + 1)
circuit.x(2)
state_vector = Statevector.from_instruction(circuit)
state_vector.draw('qsphere')
```

```python
circuit.draw('mpl')
```
