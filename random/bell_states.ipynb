```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex

def print_circuit(state):
    print("=================================")
    print("state = " + state)
    circuit = QuantumCircuit(2)
    circuit.initialize(Statevector.from_label(state).data, [0, 1])
    circuit.h(0)
    circuit.cx(0, 1)
    display(circuit.draw('mpl'))
    state_vector = Statevector.from_instruction(circuit)
    display(state_vector.draw('latex'))
    display(array_to_latex(state_vector))
    display(state_vector.draw('qsphere'))
```

```python
states = [format(x, f'0{2}b') for x in range(2 ** 2)]
print(states)
```

```python
for state in states:
    print_circuit(state)
```
